# Plotting Turkish exports from 2007 to 2016

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as mp
%matplotlib inline
```

We will load the data into our work space and convert it to a dataframe .

```python
data = pd.read_excel('Exports by country and year (top 20 country in exports).xls')
```

We need some simple data cleaning , dropping unnecessary columns . 

```python
data.drop(data.index[0:6])
```

```python
df1 = data.drop(data.index[0:6])
```

```python
df = df1.iloc[0:21,1:]
```

```python
df.head()
```

```python
index = ['Country','2016','2015','2014','2013','2012','2011','2010','2009','2008','2007']
```

```python
df.iloc[:,0] = df.iloc[:,11]
```

```python
df = df.iloc[:,0:11]
```

```python
df
```

Converting some columns to float types so that we can plot it without incidents . 

```python
df['Unnamed: 4']=df['Unnamed: 4'].astype(float)
```

```python
df['Unnamed: 10']=df['Unnamed: 10'].astype(float)
```

Changing column names 

```python
df.columns = index
```

```python
df.head()
```

```python
df.reset_index()
```

Here we prepare the data to be analyzed and plotted . 

```python
rows = [df.iloc[i].tolist() for i in range(0,21)]
```

```python
country_names = []
for row in rows :
    country_names.append(row[0])
```

```python
cb_orange = (255/255, 128/255, 14/255)

ax_list = []
for i in range(0,21):
    ax = 'ax'+str(i+1)
    ax_list.append(ax)
    
shown_x = [18,19,20]
shown_y = [0,3,6,9,12,15,18]
    
fig = plt.figure(figsize=(60,60))
plt.style.use('fivethirtyeight')

for i in range(0,21):
    for ax in ax_list:
        ax = fig.add_subplot(7,3,i+1)
        ax.plot(data.iloc[3,2:12],df.iloc[i,1:],c=cb_orange, linewidth=2)
        labels = [item.get_text() for item in ax.get_xticklabels()]
        labels = data.iloc[3,2:12]
        ax.set_xticklabels(sorted(labels))
        for tick in ax.xaxis.get_major_ticks():
            tick.label.set_fontsize(15)
        for tick in ax.yaxis.get_major_ticks():
            tick.label.set_fontsize(15)
        if i in shown_y :
            ax.tick_params(labelleft='on',size = 5)
        else :
            ax.tick_params(labelleft='off')
        if i in shown_x:
            ax.tick_params(labelbottom='on',size = 5)
        else :
            ax.tick_params(labelbottom='off')
        ax.set_title(country_names[i])
plt.show()
fig.savefig('plotting_exports.pdf')
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)
ax.plot(data.iloc[3,:][2:12],df.iloc[0,1:],c=cb_orange, linewidth=2)
ax.set_title(df.iloc[0,0])
labels = [item.get_text() for item in ax.get_xticklabels()]
labels = data.iloc[3,2:12]
ax.set_xticklabels(sorted(labels))
plt.show()
fig.savefig('plotting_german_exports.pdf')
```

# Using bokeh library to plot german exports and saving it in HTML format .

```python
from bokeh.plotting import figure,output_file,show
```

```python
p = figure(plot_width=500,plot_height=500, tools='pan,resize',logo=None)
```

```python
p.title='Germany'
p.title.text_color='Red'
p.title.text_font='times'
p.title.text_font_style='italic'
p.line(data.iloc[3,:][2:12],df.iloc[0,1:],color='red',alpha=.5)
output_file('german_exports.html')
show(p)
```
